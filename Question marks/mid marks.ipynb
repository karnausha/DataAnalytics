```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("class_marks.csv")
```

```python
df
```

```python
df.Total.value_counts()
```

```python
df.shape
```

```python
 df["Q1"] = df["Q1aM4"] + df["Q1bM6"]
 df["Q2"] = df["Q2aM6"] + df["Q2bM4"]
 df["Q3"] = df["Q3aM5"] + df["Q3bM5"]
 df["Q4"] = df["Q4aM3"] + df["Q4bM7"]
 df["Q5"] = df["Q5M10"]
 df["Q6"] = df["Q6aM4"] + df["Q6bM6"]
 df.drop(["Q1aM4", "Q1bM6","Q2aM6","Q2bM4" ,"Q3aM5" , "Q3bM5", "Q4aM3" ,"Q4bM7","Q5M10","Q6aM4","Q6bM6"], axis=1, inplace=True) 
 df
```

```python
plt.hist(df['Q5'],color='pink',edgecolor='black',bins=7)
plt.xlabel("Marks scored in Q5")
plt.ylabel("No.of students")
plt.title("Histogram of Q5")
```

# majority of students scored 5 marks,with some scoring close to the maximum

```python
df.plot()
plt.show()
```

```python
a1=df.groupby('Q1')['Total']
a1.hist(color="green",grid=False,bins=7)
plt.title('Histogram of students who scored 5-10 Total marks')
plt.xlabel("Total marks obtained")
plt.ylabel("marks in Q2")
plt.show()
```

# In Q1 major students got 6 marks

```python
a2=df.groupby('Q3')['Total']
a2.hist(color="yellow",grid=False,bins=8)
plt.title('Histogram of students who scored 35-40 Total marks')
plt.xlabel("Total marks obtained")
plt.ylabel("marks in Q3")
plt.show()
```

# the students who scored 35-40 total marks are got maximum 8 marks in Q3

```python
df.plot.scatter(x = 'Q2', y = 'Total',color='black',s=30)
plt.title("Scatter Plot for Q1")
```
